## Seperrate X and Y

```python
X = df.drop(columns=['id','booking_status'])
Y = df[['booking_status']]
```

### Preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.value_counts()
```

```python
ytrain.value_counts()/len(ytrain)
```

```python
ytest.value_counts()
```

```python
ytest.value_counts()/len(ytest)
```

### Create a neural Network

```python
xtrain.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()

# Hidden layer 1
model.add(Dense(64, activation='relu', input_dim=xtrain.shape[1]))
# Dropout Layer 1
model.add(Dropout(0.4))
# Hidden Layer 2
model.add(Dense(32, activation='relu'))
# Dropout Layer 2
model.add(Dropout(0.3))
# Output Layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

### Compile model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', 'AUC'])
```

### Train the model

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=100)
```

## Learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Cross Entropy Loss')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

### Evaluate model

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

### Predict Probability for train and test

```python
yprob_tr = model.predict(xtrain)
yprob_ts = model.predict(xtest)
```
